```python
# converting text to numerical values
from sklearn.preprocessing import LabelEncoder
columns = ['LocationAbbr','LocationDesc','Topic','TopicID','Question','QuestionID','GeoLocation','StratificationCategory1','Stratification1']
le = LabelEncoder()

for cols in columns:
    df[cols] = le.fit_transform(df[cols])
```

```python
# Dealing with outliers
df = df[(df['High_Confidence_Limit ']>5) & (df['High_Confidence_Limit ']<85)]
```

```python
# dealing with duplicates if any
df.drop_duplicates(inplace = True)
df.reset_index(inplace = True, drop = True)
```

```python
# spliiting data
x = df.drop(['Data_Value'],axis = 1)
y = df['Data_Value']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(x_train.shape,y_train.shape,x_test.shape,y_test.shape)
```

```python
# Using StandardScaler to bring data on same scale
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
# GridSearchCV to check the best model and best parameters

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import ShuffleSplit 
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
```

```python
from sklearn.model_selection import GridSearchCV

def find_best_model(x, y):
    ml = {
        'KNeighbors_Regressor': {
            'model': KNeighborsRegressor(),
            'params': {
                'n_neighbors': [3, 5, 7, 9],
                'weights': ['uniform', 'distance'],
                'p': [1, 2]
            }
        },
        'Linear_Regression': {
            'model': LinearRegression(),
            'params': {
                'fit_intercept': [True, False]
            }
        }
    }
    scores = []
    for ml_name, mp in ml.items():
        gs = GridSearchCV(mp['model'], mp['params'], cv=5, scoring='neg_mean_squared_error')
        gs.fit(x, y)
        scores.append({
            'model': ml_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })
    return pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])

find_best_model(x, y)
```

```python
# Model Building
from sklearn.linear_model import LinearRegression
lr = LinearRegression(fit_intercept=True)
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred
```
