```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Mall_Customers.csv')
df
```

```python
df.info()
```

```python
x = df.iloc[:,[3,4]].values
x
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    km = KMeans(n_clusters=i,init='k-means++',random_state=42)
    km.fit(x)
    wcss.append(km.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.title('Elbow Graph')
plt.xlabel('No. of cluster')
plt.ylabel('WCSS')
```

```python
# k = 5 bcz after 5 variations
```

```python
km = KMeans(n_clusters=5,init='k-means++',random_state=0)
y = km.fit_predict(x)
print(y)
```

```python
plt.figure(figsize=(8,8))
plt.scatter(x[y==0,0],x[y==0,1],s = 50,c = 'green',label = 'cluster-0')
plt.scatter(x[y==1,0],x[y==1,1],s = 50,c = 'red',label = 'cluster-1')
plt.scatter(x[y==2,0],x[y==2,1],s = 50,c = 'blue',label = 'cluster-2')
plt.scatter(x[y==3,0],x[y==3,1],s = 50,c = 'cyan',label = 'cluster-3')
plt.scatter(x[y==4,0],x[y==4,1],s = 50,c = 'pink',label = 'cluster-4')

plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],s= 150,label= 'centeroid',c ='black')

plt.title('Customer_Group')
plt.xlabel('Annual Income')
plt.ylabel('Spending score')
```
